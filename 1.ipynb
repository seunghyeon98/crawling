# 네이버 title, 소개글 찾기

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time

data = raw_data
data.reset_index(drop=True,inplace=True)

data['NaverTitle'] = None
data['ImageUrl'] = None
data['description'] = None



for idx in tqdm(range(len(data))):
  driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

  query =data.at[idx,'preprocessed']
  add_query = query +' 정보'
  urls = 'https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&query='+add_query
  driver.get(urls)
  # element = driver.find_element(By.XPATH, "//*[@id='newsstand']/div[1]/div/ul/li[1]/span/a[1]")

  try:
    title_elm = driver.find_element(By.XPATH,'//*[@id="main_pack"]/div[2]/div[1]/div[1]/h2/span/strong')
    title = title_elm.text
    data.at[idx,'NaverTitle'] = title
  except:
    pass

  try:
    img_elm = driver.find_element(By.XPATH,'//*[@id="main_pack"]/div[2]/div[2]/div[1]/div/div[1]/a/img')
    img = img_elm.get_attribute('src')
    data.at[idx,'ImageUrl'] = img
  except:
    pass

  try:
    description_elm = driver.find_element(By.XPATH,'//*[@id="main_pack"]/div[2]/div[2]/div[2]/div/div[2]/p')
    description = description_elm.text
    data.at[idx,'description'] = description
  except:
    pass

  

  # 웹 드라이버 종료
  driver.quit()
```

# 출연진 정보 찾기

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time
# driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

data = data_mv.head()
data.reset_index(drop=True,inplace=True)

data['감독'] =None
data['주연'] =None
data['조연'] =None


for idx in range(len(data)):
  driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

  query =data.at[idx,'preprocessed']
  add_query = '영화 ' + query +' 출연진'
  urls = 'https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&query='+add_query
  driver.get(urls)
  # element = driver.find_element(By.XPATH, "//*[@id='newsstand']/div[1]/div/ul/li[1]/span/a[1]")



  director = driver.find_element(By.XPATH,'//*[@id="main_pack"]/div[2]/div[2]/div/div/div[1]/div/div/div/ul/li/a/div/div[2]/strong').text

  
  # print(director)
  
  data.at[idx,'감독'] = director

  actor_1_list =[]
  actor_1_li = driver.find_elements(By.XPATH,'//*[@id="main_pack"]/div[2]/div[2]/div/div/div[2]/div/div/div/ul/li')
  actor_1_li_len= len(actor_1_li)

  for jdx in range(1,actor_1_li_len+1):
    actor = driver.find_element(By.XPATH,f'//*[@id="main_pack"]/div[2]/div[2]/div/div/div[2]/div/div/div/ul/li[{jdx}]/a/div/div[2]/strong').text
    actor_1_list.append(actor)


  data.at[idx,'주연'] = actor_1_list

  time.sleep(2)
  actor_2_list =[]
  actor_2_li = driver.find_elements(By.XPATH,'//*[@id="main_pack"]/div[2]/div[2]/div/div[2]/div[3]/div/div/div/ul/li')

  actor_2_li_len = len(actor_2_li)
  # print(actor_2_li_len)
  for jdx in range(1,actor_2_li_len+1):
    actor = driver.find_element(By.XPATH,f'//*[@id="main_pack"]/div[2]/div[2]/div/div[2]/div[3]/div/div/div/ul/li[{jdx}]').text
    actor_2_list.append(actor)


  data.at[idx,'조연'] = actor_2_list


  # print(actor_1_list)

  # print(actor_2_list)


  # 웹 드라이버 종료
  driver.quit()
```
